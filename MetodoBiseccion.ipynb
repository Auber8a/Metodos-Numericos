4. a. Dibuje las gr√°ficas para $ y = x^2 - 1 $ y $ y = e^{1 - x^2} $.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de valores para x
x = np.linspace(-3, 3, 400)

# Definir las funciones
y1 = x**2 - 1
y2 = np.exp(1 - x**2)

# Graficar las funciones
plt.plot(x, y1, label=r'$y = x^2 - 1$', color='blue')
plt.plot(x, y2, label=r'$y = e^{1 - x^2}$', color='red')

# A√±adir etiquetas y t√≠tulo
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gr√°ficas de $y = x^2 - 1$ y $y = e^{1 - x^2}$')

# A√±adir leyenda y mostrar la gr√°fica
plt.legend()
plt.grid()
plt.show()
```
