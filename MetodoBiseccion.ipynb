4. a. Dibuje las gráficas para $ y = x^2 - 1 $ y $ y = e^{1 - x^2} $.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de valores para x
x = np.linspace(-3, 3, 400)

# Definir las funciones
y1 = x**2 - 1
y2 = np.exp(1 - x**2)

# Graficar las funciones
plt.plot(x, y1, label=r'$y = x^2 - 1$', color='blue')
plt.plot(x, y2, label=r'$y = e^{1 - x^2}$', color='red')

# Añadir etiquetas y título
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráficas de $y = x^2 - 1$ y $y = e^{1 - x^2}$')

# Añadir leyenda y mostrar la gráfica
plt.legend()
plt.grid()
plt.show()
```
