## Usando el metodo de Newton

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci√≥n y su derivada
def f(x):
    return x**3 - 3*x**2 + x - 1

def f_prime(x):
    return 3*x**2 - 6*x + 1

# Crear un rango de valores de x para la gr√°fica
x = np.linspace(-2, 4, 400)  # Ajusta el rango seg√∫n sea necesario
y = f(x)

# Graficar la funci√≥n
plt.plot(x, y, label="y = x^3 - 3x^2 + x - 1")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gr√°fica de y = x^3 - 3x^2 + x - 1')
plt.grid(True)
plt.legend()
plt.show()

# M√©todo de Newton
def newton_method(x0=-8, tol=1e-6, max_iter=100):
    points = []
    for i in range(max_iter):
        y = f(x0)
        y_prime = f_prime(x0)
        points.append((x0, y))  # Guardar el punto (x, f(x)) en cada iteraci√≥n

        if abs(y) < tol:
            print(f"Ra√≠z encontrada: {x0} en {i+1} iteraciones.")
            return x0, points

        if y_prime == 0:
            print("Derivada cero, el m√©todo falla.")
            return None, points

        x0 = x0 - y / y_prime  # Aplicaci√≥n del m√©todo de Newton

    print("M√°ximo de iteraciones alcanzado sin convergencia.")
    return None, points

# Ejemplo de uso con un valor inicial
x0 = 100  # Ajusta el valor inicial seg√∫n sea necesario
raiz, puntos = newton_method(x0)

# Imprimir los puntos evaluados durante las iteraciones
for x, y in puntos:
    print(f"f({x}) = {y}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci√≥n y su derivada
def f(x):
    return x**3 - 3*x**2 + x - 1

def f_prime(x):
    return 3*x**2 - 6*x + 1

# Crear un rango de valores de x para la gr√°fica
x = np.linspace(-2, 4, 400)  # Ajusta el rango seg√∫n sea necesario
y = f(x)

# M√©todo de Newton
def newton_method(x0, tol=1e-6, max_iter=100):
    points = []
    for i in range(max_iter):
        y_val = f(x0)
        y_prime = f_prime(x0)
        points.append((x0, y_val))  # Guardar el punto (x, f(x)) en cada iteraci√≥n

        if abs(y_val) < tol:
            print(f"Ra√≠z encontrada: {x0} en {i+1} iteraciones.")
            return x0, points

        if y_prime == 0:
            print("Derivada cero, el m√©todo falla.")
            return None, points

        x0 = x0 - y_val / y_prime  # Aplicaci√≥n del m√©todo de Newton

    print("M√°ximo de iteraciones alcanzado sin convergencia.")
    return None, points

# Valores iniciales para probar
initial_guesses = [2, 3, -1]  # x0 = 2, x0 = 3 y x0 = -1

# Guardar los resultados
results = []

for x0 in initial_guesses:
    x_r, points = newton_method(x0)
    results.append((x0, x_r, points))
```

```python
# Graficar la funci√≥n y la ra√≠z encontrada para cada x0
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="$x^3 - 3x^2 + x - 1$")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)

# Marcar las ra√≠ces encontradas en cada uno de los intentos
for x0, x_r, _ in results:
    if x_r is not None:
        plt.scatter(x_r, f(x_r), label=f"Ra√≠z (x0={x0})", zorder=5)

plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.title("Plot of the equation $x^3 - 3x^2 + x - 1$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
# Graficar los puntos x durante las iteraciones para cada x0
plt.figure(figsize=(8, 6))

for x0, _, points in results:
    x_points = [point[0] for point in points]
    plt.plot(x_points, "o-", label=f"x points for x0 = {x0}")

plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.title("Plot of x points during iterations")
plt.xlabel("Index")
plt.ylabel("x value")
plt.legend()
plt.show()
```
