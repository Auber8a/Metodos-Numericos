###  <span style="color:orange"> *1. Sea $( f(x) = -x^3 - \cos x )$ y $( p_0 = -1 )$. Use el método de Newton y de la Secante para encontrar $( p_2 )$. ¿Se podría usar $( p_0 = 0 )$?* </span>.    


### - Método de Newton   
$$f(x) = -x^3 - \cos{x}$$   
$$f'(x) = -3x^2 + \sin{x}$$  
     
$$x_n = x_{n-1} - \frac{f(x_{n-1})}{f'(x_{n-1})}$$    

### - Aplicamos el método para obtener:
$x_n = -0.865$     
Dado que al evaluar en la función f'(0), el denominador seria **cero**. Por lo cual no se podria aplicar el metodo de Newton con $p_0 = 0$.   

### - Método de la Secante
$p_0 = -1$, usamos un valor cercano a cero. Usamos $p_0 = 0.5$     
$$x_n = x_{n-1} - y_{n-1} * \frac{x_{n-1} - x_{n-2}}{y_{n-1} - y_{n-2}}$$    

### - Aplicamos el método para obtener:    
$x_n = -0.865$    
No es posible usar $p_0 = 0$, debido a que el siguiente valor seria $p_1 = -1$ y ese no seria un bien punto de partida.





```python
import numpy as np

def f(x):
    return -x**3 - np.cos(x)

def df(x):
    return -3 * x**2 + np.sin(x)


def newton_method(f, df, p0, tol=1e-6, max_iter=100):
    p = p0
    for i in range(max_iter):
        p_next = p - f(p) / df(p)
        if abs(p_next - p) < tol:
            return p_next, i + 1  
        p = p_next
    return None, max_iter  


def secant_method(f, p0, p1, tol=1e-6, max_iter=100):
    for i in range(max_iter):
        if abs(f(p1) - f(p0)) < tol:
            return None, i  
        p_next = p1 - f(p1) * (p1 - p0) / (f(p1) - f(p0))
        if abs(p_next - p1) < tol:
            return p_next, i + 1  
        p0, p1 = p1, p_next
    return None, max_iter  


p0 = -1


sol_newton, iter_newton = newton_method(f, df, p0)
print(f"Método de Newton: p2 = {sol_newton}, iteraciones = {iter_newton}")


p1 = -0.5
sol_secant, iter_secant = secant_method(f, p0, p1)
print(f"Método de la Secante: p2 = {sol_secant}, iteraciones = {iter_secant}")
```

### <span style="color:orange"> *d. $( x - 0.8 - 0.2 \sin x = 0 ), \quad ([0, \pi / 2])$* </span>   
### - Método de Newton   
$p_0 = \pi / 4$
$$f(x) = x - 0.8 - 0.2 \sin (x)$$   
$$f'(x) = 1 - \cos (x)$$  
     
$$x_n = x_{n-1} - \frac{f(x_{n-1})}{f'(x_{n-1})}$$    

### - Aplicamos el método para obtener:
$$x_n = 0.964$$     

### - Método de la Secante
$p_0 = 0$, $p_1 = \pi / 2$     
$$x_n = x_{n-1} - y_{n-1} * \frac{x_{n-1} - x_{n-2}}{y_{n-1} - y_{n-2}}$$    

### - Aplicamos el método para obtener:    
$$x_n = 0.964$$    


```python
import numpy as np
from scipy.optimize import newton

def f1(x):
    return x**3 - 2 * x**2 - 5

def f1_prime(x):
    return 3 * x**2 - 4 * x

def f2(x):
    return x**3 + 3 * x**2 - 1

def f2_prime(x):
    return 3 * x**2 + 6 * x

def f3(x):
    return x - np.cos(x)

def f3_prime(x):
    return 1 + np.sin(x)

def f4(x):
    return x - 0.8 - 0.2 * np.sin(x)

def f4_prime(x):
    return 1 - 0.2 * np.cos(x)


def metodo_secante(f, x0, x1, tol=1e-4, max_iter=100):
    for _ in range(max_iter):
        if abs(f(x1) - f(x0)) < tol:
            break
        x_temp = x1 - f(x1) * (x1 - x0) / (f(x1) - f(x0))
        x0, x1 = x1, x_temp
        if abs(f(x1)) < tol:
            break
    return x1


def metodo_newton(f, f_prime, x0, tol=1e-4, max_iter=100):
    raiz = newton(f, x0, fprime=f_prime, tol=tol, maxiter=max_iter)
    return raiz


raiz_f1_secante = metodo_secante(f1, 1, 4)
raiz_f1_newton = metodo_newton(f1, f1_prime, x0=2.5)

raiz_f2_secante = metodo_secante(f2, -3, -2)
raiz_f2_newton = metodo_newton(f2, f2_prime, x0=-2.5)

raiz_f3_secante = metodo_secante(f3, 0, np.pi/2)
raiz_f3_newton = metodo_newton(f3, f3_prime, x0=np.pi/4)

raiz_f4_secante = metodo_secante(f4, 0, np.pi/2)
raiz_f4_newton = metodo_newton(f4, f4_prime, x0=np.pi/4)


print("Resultados para las funciones:")
print(f"1. f1(x): Secante = {raiz_f1_secante:.4f}, Newton = {raiz_f1_newton:.4f}")
print(f"2. f2(x): Secante = {raiz_f2_secante:.4f}, Newton = {raiz_f2_newton:.4f}")
print(f"3. f3(x): Secante = {raiz_f3_secante:.4f}, Newton = {raiz_f3_newton:.4f}")
print(f"4. f4(x): Secante = {raiz_f4_secante:.4f}, Newton = {raiz_f4_newton:.4f}")
```

### <span style="color:orange"> *4. El polinomio de cuarto grado $$[f(x) = 230x^4 + 18x^3 + 9x^2 - 221x - 9]$$* </span> 

### <span style="color:orange"> *tiene dos ceros reales, uno en $([-1,0])$ y el otro en $([0,1])$. Intente aproximar estos ceros dentro de $(10^{-6})$ con:* </span> 

### <span style="color:orange"> *a. El método de la secante (use los extremos como las estimaciones iniciales)* </span> 

En el intervalo [−1,0]:   

$p_0 = -1$, $p_1 = 0$
$$x_n1 = -0.04065$$  

En el intervalo [0,1]:    
$p_2 = 0$, $p_3 = 1$
$$x_n2 = -0.04065$$       
$$x_n = x_{n-1} - y_{n-1} * \frac{x_{n-1} - x_{n-2}}{y_{n-1} - y_{n-2}}$$  

---
### <span style="color:orange"> *b. El método de Newton (use el punto medio como estimación inicial)* </span> 

En el intervalo [−1,0]:   

$p_0 = -1$, $p_1 = 0$
$$x_n1 = -0.04065$$  

En el intervalo [0,1]:    
$p_2 = 0$, $p_3 = 1$
$$x_n2 = -0.04065$$       
$$x_n = x_{n-1} - \frac{f(x_{n-1})}{f'(x_{n-1})}$$ 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton

def f(x):
    return 230 * x**4 + 18 * x**3 + 9 * x**2 - 221 * x - 9

def f_prime(x):
    return 920 * x**3 + 54 * x**2 + 18 * x - 221


def metodo_newton(f, f_prime, x0, tol=1e-6, max_iter=100):
    raiz = newton(f, x0, fprime=f_prime, tol=tol, maxiter=max_iter)
    return raiz


def metodo_secante(f, x0, x1, tol=1e-6, max_iter=100):
    for _ in range(max_iter):
        if abs(f(x1) - f(x0)) < tol:
            break
        x_temp = x1 - f(x1) * (x1 - x0) / (f(x1) - f(x0))
        x0, x1 = x1, x_temp
        if abs(f(x1)) < tol:
            break
    return x1

# Aproximación de los ceros
# [-1, 0]
raiz_secante_1 = metodo_secante(f, -1, 0)
raiz_newton_1 = metodo_newton(f, f_prime, x0=(-1 + 0) / 2)

# [0, 1]
raiz_secante_2 = metodo_secante(f, 0, 1)
raiz_newton_2 = metodo_newton(f, f_prime, x0=(0 + 1) / 2)

print("Cero en [-1, 0]:")
print(f"  Método de la secante: {raiz_secante_1:.6f}")
print(f"  Método de Newton: {raiz_newton_1:.6f}")

print("\nCero en [0, 1]:")
print(f"  Método de la secante: {raiz_secante_2:.6f}")
print(f"  Método de Newton: {raiz_newton_2:.6f}")

x = np.linspace(-1.5, 1.5, 1000)
y = f(x)

plt.figure(figsize=(10, 6))
plt.axhline(0, color='black', linestyle='--', linewidth=0.8)
plt.plot(x, y, label="f(x)", color="orange")
plt.scatter([raiz_secante_1, raiz_secante_2], [0, 0], color="red", label="Raíces (Secante)")
plt.scatter([raiz_newton_1, raiz_newton_2], [0, 0], color="green", marker='x', label="Raíces (Newton)")
plt.title("Raíces del polinomio usando Newton y la Secante")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
